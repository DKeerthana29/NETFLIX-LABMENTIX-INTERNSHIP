## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ── Data Wrangling Pipeline ──────────────────────────────────────────────

# Step 1: Merge titles and credits datasets on common 'id' column
df = pd.merge(titles, credits, on='id', how='left')
print('Shape after merge:', df.shape)

# Step 2: Remove duplicate rows
df = df.drop_duplicates()
print('Shape after removing duplicates:', df.shape)

# Step 3: Drop rows where imdb_score is missing (key analysis variable)
df = df.dropna(subset=['imdb_score'])
print('Shape after dropping missing imdb_score:', df.shape)

# Step 4: Fill remaining missing values with sensible defaults
df['runtime'].fillna(df['runtime'].median(), inplace=True)  # Use median for runtime
df['genres'].fillna('Unknown', inplace=True)                # Label missing genres
df['age_certification'].fillna('Not Rated', inplace=True)   # Label missing certifications
df.fillna(0, inplace=True)                                  # Fill other NaNs with 0

# Step 5: Correct data types
df['release_year'] = df['release_year'].astype(int)

# Step 6: Preview the cleaned dataset
print('Final cleaned dataset shape:', df.shape)
display(df.head())
````

### What all manipulations have you done and insights you found?

The following data preprocessing steps were performed:

1. **Merging:** The titles and credits datasets were joined on the `id` column using a left join to preserve all title records.
2. **Deduplication:** Duplicate rows were removed to ensure data integrity.
3. **Missing value handling:** Rows with missing `imdb_score` were dropped since it is the primary analysis variable. Runtime was imputed with the median; genre and certification were filled with descriptive labels.
4. **Data type correction:** `release_year` was cast to integer for proper grouping.

**Key observations from wrangling:**
- The merged dataset is richer, combining movie metadata with cast/crew details.
- IMDb score is available for a large majority of records.
- Genre and certification fields had notable missingness, handled with informative labels to avoid data loss.

#### Chart - 20 : IMDb Score vs Votes vs Popularity (Bubble Chart — Multivariate)

````python
# Chart 20 – Multivariate: Bubble chart — IMDb score vs votes, bubble size = popularity
plt.figure(figsize=(12, 6))
sample = df.dropna(subset=['imdb_score', 'imdb_votes', 'tmdb_popularity']).sample(
    min(500, len(df)), random_state=42)

# Normalize bubble size for visibility
bubble_size = (sample['tmdb_popularity'] / sample['tmdb_popularity'].max()) * 300 + 10

scatter = plt.scatter(
    sample['imdb_votes'], sample['imdb_score'],
    s=bubble_size, alpha=0.4, c=sample['tmdb_popularity'],
    cmap='plasma', edgecolors='none')

plt.colorbar(scatter, label='TMDB Popularity')
plt.title('IMDb Score vs IMDb Votes (Bubble Size = TMDB Popularity)', fontsize=14)
plt.xlabel('IMDb Votes (Number of Ratings)', fontsize=12)
plt.ylabel('IMDb Score', fontsize=12)
plt.xscale('log')  # Log scale to handle wide vote range
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?
A bubble chart encodes three variables simultaneously: X-axis (votes), Y-axis (score), and bubble size/colour (popularity). This is a powerful multivariate chart that reveals whether high-vote, high-score content also tends to be popular — and how the three metrics relate.

##### 2. What is/are the insight(s) found from the chart?
Titles with more IMDb votes tend to have more stable scores (clustering in the 7–8 range), confirming that well-known titles converge toward consensus ratings. Highly popular titles (larger/brighter bubbles) are spread across a range of scores — popularity alone does not predict rating. A few large bubbles in the high-vote, high-score quadrant represent blockbuster titles that are both critically and commercially successful.

##### 3. Will the gained insights help creating a positive business impact?
Yes — this chart identifies the 'sweet spot' of high votes + high score + high popularity, which represents the most valuable content category. Targeting acquisitions or productions in this quadrant maximizes multi-dimensional value. Negatively, the chart also shows many high-popularity titles with mediocre scores, indicating that viral marketing can temporarily inflate popularity without translating to lasting quality perception — a risk for brand credibility.
